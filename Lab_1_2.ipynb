# Завдання 2

Побудуйте графік поверхні (3D surface plot) для візуалізації залежності
між трьома числовими змінними з використанням бібліотеки Seaborn.
Налаштуйте відображення графіка, включаючи назву графіку, підписи
осей і інші необхідні параметри для кращого розуміння візуалізації.
Додайте кольорову легенду.


```python
from mpl_toolkits.mplot3d import Axes3D

# Створення даних для x, y та z
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Створення ділянки 3D-поверхні
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#Побудова графіку поверхні
surface = ax.plot_surface(x, y, z, cmap='rocket', edgecolor='none')

# Додавання заголовків
ax.set_xlabel('Вісь X')
ax.set_ylabel('Вісь Y')
ax.set_zlabel('Вісь Z')
ax.set_title('3D поверхневий графік')

# Додавання смуги кольорів, яка відображає значення
fig.colorbar(surface, shrink=0.5, aspect=5)

plt.show()
```
