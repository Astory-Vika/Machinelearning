```python
# Зчитуємо дані з вказаного URL
url = "https://uk.wikipedia.org/wiki/%D0%A1%D0%B5%D1%80%D0%B5%D0%B4%D0%BD%D1%8F_%D0%B7%D0%B0%D1%80%D0%BE%D0%B1%D1%96%D1%82%D0%BD%D0%B0_%D0%BF%D0%BB%D0%B0%D1%82%D0%B0_%D0%B2_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D1%96"

# Зчитуємо таблицю в DataFrame
data = pd.read_html(url, header=0)[2]  # Обираємо другий елемент списку, де знаходиться таблиця
data
```

6.	Видаліть передостанній і останній стовпчики ( Кількість кгхліба[20] на середнюзарплату,  Кільксть БігМаків на середню зарплату). Далі працюйте з оновленою таблицею. 

```python
data = data.drop(columns=["Кількість кг хліба[20] на середню зарплату", "Кількість БігМаків на середню зарплату"])
data
```

7.	Замінити тип нечислових даних на числові.

```python
# Замінимо "," на "."
data['В національній валюті'] = data['В національній валюті'].str.replace(',', '.')
data['В доларах США'] = data['В доларах США'].str.replace(',', '.')

#Видалимо трикутники
data["В національній валюті"] = data["В національній валюті"].str.replace('▲', "").str.replace('▼', "")
data["В доларах США"] = data["В доларах США"].str.replace('▲', "").str.replace('▼', "")

#Видалимо * в 2022* і 2023*
data["Рік"] = data["Рік"].str.replace("*", "")
# Перетворимо дані у числовий формат
data['В національній валюті'] = data['В національній валюті'].astype(float)
data['В доларах США'] = data['В доларах США'].astype(float)
data["Рік"] = data["Рік"].astype(int)
data
```

```python
#Ще раз перевіримо тип даних

data.dtypes
```

8.	Порахуйте частку пропусків в кожній колонці (за допомогою методів isnull , sum, len)

```python
sumIsNull = pd.isnull(data)
print(sumIsNull)
sum1 = 0
sum2 = 0
sum3 = 0
for i in range(1, len(data)):
    if (sumIsNull["Рік"][i] != False):
        sum1 = sum1 + 1
    if (sumIsNull["В національній валюті"][i] != False):
        sum2 = sum2 + 1
    if (sumIsNull["В доларах США"][i] != False):
        sum3 = sum3 + 1
        
print("Частка пропусків за стовпчиком \"Рік\" становить", sum1/len(data))
print("Частка пропусків за стовпчиком \"В національній валюті\" становить", sum2/len(data))
print("Частка пропусків за стовпчиком \"В доларах США\" становить", sum3/len(data))
```

9.	Замініть всі пропущені значення на середні значення по (стовпцю).

```python
data['Рік']=data['Рік'].fillna(data['Рік'].mean())
data['В національній валюті']=data['В національній валюті'].fillna(data['В національній валюті'].mean())
data['В доларах США']=data['В доларах США'].fillna(data['В доларах США'].mean())
data
```
